# Rolling and Shifting

We'll shift the data forward, so we're not incorporating information from the current day into our predictor. If we didn't use shift, then the algorithm would have knowledge of the actual target.

Find the 7-day rolling sum of the target. If the price went up for all 7 days, this would be 7. If it went up 0 days, this would be 0.

```python
data_MSFT
```

```python
data_AAPL
```

```python
data_GOOG
```

```python
weekly_trend_MSFT = data_MSFT.shift(1).rolling(7).sum()["Target_MSFT"]
weekly_trend_AAPL = data_AAPL.shift(1).rolling(7).sum()["Target_AAPL"]
weekly_trend_GOOG = data_GOOG.shift(1).rolling(7).sum()["Target"]
```

```python
data_MSFT["weekly_mean_MSFT"] = weekly_mean_MSFT / data_MSFT["Actual_Close_MSFT"]
data_MSFT["quarterly_mean_MSFT"] = quarterly_mean_MSFT / data_MSFT["Actual_Close_MSFT"]
data_MSFT["annual_mean_MSFT"] = annual_mean_MSFT / data_MSFT["Actual_Close_MSFT"]
```

```python
data_AAPL["weekly_mean_AAPL"] = weekly_mean_AAPL / data_AAPL["Actual_Close_AAPL"]
data_AAPL["quarterly_mean_AAPL"] = quarterly_mean_AAPL / data_AAPL["Actual_Close_AAPL"]
data_AAPL["annual_mean_AAPL"] = annual_mean_AAPL / data_AAPL["Actual_Close_AAPL"]
```

```python
data_GOOG["weekly_mean_GOOG"] = weekly_mean_GOOG / data_GOOG["Actual_Close_GOOG"]
data_GOOG["quarterly_mean_GOOG"] = quarterly_mean_GOOG / data_GOOG["Actual_Close_GOOG"]
data_GOOG["annual_mean_GOOG"] = annual_mean_GOOG / data_GOOG["Actual_Close_GOOG"]
```
