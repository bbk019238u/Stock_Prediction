# APPLE STOCK DATA

```python
DATA_PATH = "AAPL_data.json"

if os.path.exists(DATA_PATH):
    # Read from file if we've already downloaded the data.
    with open(DATA_PATH) as f:
        AAPL_Hist = pd.read_json(DATA_PATH)
else:
    AAPL = yf.Ticker("AAPL")
    AAPL_Hist = AAPL.history(period="max")

    # Save file to json in case we need it later. Do not to re-download everytime.
    AAPL_Hist.to_json(DATA_PATH)
```

# DATA PREP FOR ML

```python
#MSFT

# Ensure we know the actual closing price
data_MSFT = MSFT_Hist[["Close_MSFT"]]
data_MSFT = data_MSFT.rename(columns = {'Close_MSFT':'Actual_Close_MSFT'})

# Setup our target.  This identifies if the price went up or down
data_MSFT["Target_MSFT"] = MSFT_Hist.rolling(2).apply(lambda x: x.iloc[1] > x.iloc[0])["Close_MSFT"]
```

```python
#AAPL

# Ensure we know the actual closing price
data_AAPL = AAPL_Hist[["Close_AAPL"]]
data_AAPL = data_AAPL.rename(columns = {'Close_AAPL':'Actual_Close_AAPL'})

# Setup our target.  This identifies if the price went up or down
data_AAPL["Target_AAPL"] = AAPL_Hist.rolling(2).apply(lambda x: x.iloc[1] > x.iloc[0])["Close_AAPL"]
```

```python
#GOOG

# Ensure we know the actual closing price
data_GOOG = GOOG_Hist[["Close_GOOG"]]
data_GOOG = data_GOOG.rename(columns = {'Close_GOOG':'Actual_Close_GOOG'})

# Setup our target.  This identifies if the price went up or down
data_GOOG["Target"] = GOOG_Hist.rolling(2).apply(lambda x: x.iloc[1] > x.iloc[0])["Close_GOOG"]
```

# CREATING THE TRAINING DATA 

```python
Stock_data = Stock_Hist[["Close_GOOG", "Close_AAPL", "Close_MSFT"]]
Stock_data = Stock_data.rename(columns = {'Close_GOOG':'Actual_Close_GOOG', 'Close_AAPL':'Actual_Close_AAPL', 'Close_MSFT':'Actual_Close_MSFT'})

# Setup our target.  This identifies if the price went up or down
Stock_data[["Target_GOOG", "Target_AAPL", "Target_MSFT"]] = Stock_Hist.rolling(2).apply(lambda x: x.iloc[1] > x.iloc[0])[["Close_GOOG", "Close_AAPL", "Close_MSFT"]]
```

```python
predictors = ["Close_AAPL", "Volume_AAPL", "Open_AAPL", "High_AAPL", "Low_AAPL", "Close_MSFT", "Volume_MSFT", "Open_MSFT", "High_MSFT", "Low_MSFT", "Close_GOOG", "Volume_GOOG", "Open_GOOG", "High_GOOG", "Low_GOOG"]
Stock_data = Stock_data.join(Stocks_prev[predictors]).iloc[1:]
```

```python
Stock_data
```

```python
Stock_data.plot.line(y=["Actual_Close_AAPL", "Actual_Close_MSFT", "Actual_Close_GOOG"], use_index=True)
```

# CREATING ML MODEL

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np
```

```python
Model = RandomForestClassifier(n_estimators=300, criterion="entropy", min_samples_leaf=250, random_state=1)
```

```python
train = Stock_data.iloc[:-100]
test = Stock_data.iloc[-100:]

Model.fit(train[predictors], train[["Target_GOOG", "Target_AAPL", "Target_MSFT"]])
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
preds = Model.predict(test[predictors])
preds = pd.DataFrame(preds, index=test.index)
```
